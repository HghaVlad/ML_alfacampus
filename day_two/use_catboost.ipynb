```python
import catboost as cb
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
my_xtrain = pd.read_csv("E:\Code works\Work\Python\Works\max_flask_app\\alfa\my_big_featurex.csv")
my_ytrain = pd.read_csv("E:\Code works\Work\Python\Works\max_flask_app\\alfa\my_big_featurey.csv")
```

```python
x_train, x_test, y_train, y_test = train_test_split(my_xtrain, my_ytrain, test_size=0.1)
```

```python
dataset = cb.Pool(x_train, y_train['flag'])
```

```python
model2 = cb.CatBoostRegressor(loss_function="RMSE")
grid = {'iterations': [190, 180],
        'learning_rate': [0.03, 0.1],
        'depth': [4, 5, 6,],
        'l2_leaf_reg': [0.2, 1, 3]}
model2.grid_search(grid, dataset)
```

```python
pred = model2.predict(x_test)
rmse = (np.sqrt(mean_squared_error(y_test['flag'], pred)))
r2 = r2_score(y_test['flag'], pred)
score = model2.score(x_test, y_test['flag'])
local_score = model2.score(x_train, y_train['flag'])
print("Testing performance")
print("RMSE: {:.2f}".format(rmse))
print("R2: {:.2f}".format(r2))
print("Score: {:.4f}".format(score))
print("Local Score: {:.4f}".format(local_score))
```
